# Update sequences in Postgres

## 1. Get list of plates in final workflow or experiment

```python
exp_workflow_barcode_list = ['WF10125','WF10128','WF10130','WF10132', 'WF10134', 'WF10136', 'WF10139', 'WF10141', 'WF10144', 'WF10149', 'WF10157', 'WF10158', 'WF10163', 'WF10169', 'WF10173', 'WF10181'] # ['WF10179'] #
plate_list = None # 
plate_types_to_fetch = ['lcms_final'] # ['hts'] # 

if exp_workflow_barcode_list is not None:
    plate_type_list = []
    plate_list = []
    for str_find in plate_types_to_fetch:
        for exp_workflow_barcode in exp_workflow_barcode_list:
            if 'EXP' in exp_workflow_barcode:
                plate_list_subset = plates_from_exp(exp_workflow_barcode, str_find)
            elif 'WF' in exp_workflow_barcode:
                plate_list_subset = plates_from_workflow(exp_workflow_barcode, str_find)
            plate_list += plate_list_subset
            plate_type_list += [str_find]*len(plate_list_subset)
else:
    plate_type_list = plate_types_to_fetch * len(plate_list)

print(f"Plate list ({len(plate_list)}): {[(p, pt) for p, pt in zip(plate_list,plate_type_list)]}")
```

## 2. Run library update to get library on plate into Postgres table and S3 datasets

```python
## update sequences in postgres
plate_type_final = plate_type_list[-1]

# HTS plate
if plate_type_final=='hts':
    library_update = DetectionsLibraryUpdate()
    library_update.get_library_update(plate_list, plate_type_list, plate_type_final)

# LCMS plate
elif plate_type_final=='lcms_final':
    library_update = DetectionsLibraryUpdate()
    library_update.get_library_update(plate_list, plate_type_list, plate_type_final)
```
